# 運用 pyhanlp 將 ptt 日旅版的文章中的景點挑出來

```python
from pyhanlp import *
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.tokenize import TreebankWordTokenizer
from nltk.tag import pos_tag
from nltk import ne_chunk, Tree

def change_to_nltk_style(a):
    newlist=[]
    for i in a:
        newlist.append((i.word,str(i.nature)))
    return newlist
```

### 把預先爬好的文章的.txt檔讀出來

```python
s=""
for i in range(10,55):
    f1=str(i)+".txt"
    try:
        with open(f1,"r") as file:
            s=s+"\n這個是分隔\n"+file.read()
    except:
        print("no file")
        

strr=HanLP.convertToSimplifiedChinese(s)
```

```python
l1=HanLP.segment(HanLP.convertToSimplifiedChinese(s))
```

```python
newL=change_to_nltk_style(l1)
```

```python
def get_sp_type(L,need=[]):
    re=[]
    for i in need:
        for j in L:
            if(j[1]==i):
                re.append(j[0])
    return re
```

### 挑出地名

```python
get_sp_type(newL,["ns","nsf"])  
```
